````python
from langchain_google_genai import ChatGoogleGenerativeAI
import os
from dotenv import load_dotenv
import re
````

````python
load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")
````

````python
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash",google_api_key=api_key)
output = llm.invoke("Write code to print pallindrome in c")
````

````python
code_match = re.search(r"```(?:\w+)?\n([\s\S]*?)```", output.content)
code_only = code_match.group(1) if code_match else output.content
````

````python
print(code_only)
````
